```python
api_url = "https://custom.resbank.co.za/SarbWebApi/MCM/Contributions/VALRATES"
response = requests.get(api_url)
if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
df = pd.DataFrame(data)
df[:]
```

```python
xml_data = df["XMLData"].values[0]
print(xml_data)
```

```python
print(type(xml_data))
```

```python
root = ET.fromstring(xml_data)
print(root)
```

```python
# Function to recursively print elements
def print_elements(element, indent=""):
    print(f"{indent}Tag: {element.tag}, Text: {element.text}")
    for child in element:
        print_elements(child, indent + "  ")
```

```python
# Start the recursive printing from the root element
print_elements(root)
```

```python
# Iterate through the elements and extract Column and Text values
data = []

for line in root.findall(".//Line"):
    columns = line.findall("./Column")
    column_values = [col.text if col.text else "None" for col in columns]
    data.append(column_values)

print(data)
```

```python
# Create a DataFrame from the extracted data
df = pd.DataFrame(data, columns=["Column", "Text"])

# Filter out rows where "None"
df = df[~df['Text'].isin(['None', '0.00'])]

# Reset the index
df.reset_index(drop=True, inplace=True)

# Print the DataFrame
print(df)
```

```python
# Get the current date
current_date = datetime.date.today()

# Calculate the date of the previous working day
if current_date.weekday() in {0, 5, 6}:  # Monday, Saturday, or Sunday
    previous_working_day = current_date - datetime.timedelta(days=3)
else:
    previous_working_day = current_date - datetime.timedelta(days=1)

# Format the date as YYYYMMDD
date_str = previous_working_day.strftime('%Y%m%d')

# Generate the CSV file name
csv_file_name = f'{date_str}_Rates.csv'

# Save the DataFrame to the CSV file
df.to_csv(csv_file_name)
```
